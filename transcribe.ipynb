```python
pip install openai-whisper
import whisper

model = whisper.load_model("medium")

result = model.transcribe("commentary.wav", language="en")

with open("commentary_full.txt", "w", encoding="utf-8") as f:
    f.write(result["text"])

print("Full transcript saved to commentary_full.txt")

# Save segmented transcript with timestamps
with open("commentary_segments.txt", "w", encoding="utf-8") as f:
    for segment in result["segments"]:
        start = segment['start']
        end = segment['end']
        text = segment['text']
        f.write(f"[{start:.2f}s -> {end:.2f}s] {text}\n")

print("Segmented transcript saved to commentary_segments.txt")
```
